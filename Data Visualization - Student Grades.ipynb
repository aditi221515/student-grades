```python
import pandas as pd
import numpy as np
```

```python
### matplotlib.pyplot is a 2D plotting library that creates high-quality figures

from matplotlib import pyplot as plt

### this command allows you to view your plots in the notebook

%matplotlib inline
```

```python
### seaborn is complimentary to matplotlib and allows for further customization 

import seaborn as sns

### set the color palette for all the plots in the notebook (optional)

sns.set_palette("Set2")
```

```python
## import EduGradeData dataset ##

edu = pd.read_csv("EduGradeData.csv")
edu.head()
```

```python
edu.drop(columns="home_state", inplace = True)
```

```python
edu.drop(columns="level_of_study", inplace = True)
```

```python
edu.head()
```

```python
## Histograms allow to visualize the distribution (how spread out the values are) of a numeric variable for one or several groups. 
## The values are split in bins, each bin is represented as a bar.

## A simple histogram to show the distribution of column 'grade' 

sns.histplot(edu["grade"], kde = True)
```

```python
## Density plots allow to visualize the distribution of a numeric variable for one or several groups
## They are very well adapted for large datasets

### A simple density plot to show the distribution of column 'grade'

sns.kdeplot(edu['grade'], shade = True, color = "teal")
```

```python
### plotting two density plots in the same figure 

plt.figure(figsize=(15,5)) ## change the figure size

sns.kdeplot(edu['hours'], shade = True, color = "blue", label = "hours")
sns.kdeplot(edu['exercise'], shade = True, color = "red", label = "exercise")

plt.legend() ## add a legend 
plt.show() ## show both plots in the same figure 
```

```python
## A boxplot summarizes the distribution of a numeric variable for one or several groups. 
## It allows to quickly get the median, quartiles and outliers but also hides the dataset individual data points.

### Creating a boxplot to show the distribution of column 'grade'

sns.boxplot(y = edu["grade"])
```

```python
### creating a series of boxplots to show the distribution of grades across groups

sns.boxplot(x = edu["exercise"], y = edu["grade"])
```

```python
### creating a series of boxplots to show the distribution of grades across groups

sns.boxplot(x = edu["hours"], y = edu["grade"])
```

```python
## A scatter plot displays the relationship between 2 numeric variables
## Each data point is represented as a circle
## Scatterplots will commonly include a trendline to help make the relationship between the variables clearer

### Create a scatterplot to show the linear relationship between 'grade' and 'hours'

plt.figure(figsize=(10,10))

sns.regplot(x = edu["grade"], y = edu["hours"], color = "black", fit_reg = False)

plt.grid()
```

```python
### create a scatterplot to show the relationship between 'grade' and 'hours' by 'age'

sns.lmplot(data = edu, x = 'grade', y = 'hours', fit_reg = False, hue = "age")

plt.grid()
```

```python
## A barplot shows the relationship between a numeric and a categoric variable
## Each entity of the categoric variable is represented as a bar
## The size of the bar represents its numeric value

### Creating a simple barplot to show the average 'grade' across 'gender'

sns.barplot(data = edu, x = "gender", y = "grade", ci = None)

plt.title('Grade Statistic by Student Gender')
plt.grid()
```

```python
### Create a barplot to show average grade across students' gender

plt.figure(figsize=(20,5))

sns.barplot(data = edu, x = "gender", y = "grade", ci = None)

plt.xticks(rotation = 45)
plt.title('Attack Statistic by Student Gender')
```

```python
### create a barplot showing the average grade by level of fitness across gender

plt.figure(figsize=(20,10))

sns.barplot(data = edu, x = "level_of_fit", y = "grade", hue = "gender", ci = None)

plt.xticks(rotation = -45)
plt.title('Grade Statistic by Level of Fitness and Gender')
```
